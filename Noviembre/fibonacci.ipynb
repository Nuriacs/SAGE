## Fibonacci

### Número m-ésimo de Fibonacci recursivamente

```python
def fib_rec(m):
    if m==0:
        return 0
    elif m==1:
        return 1
    else:
        return fib_rec(m-1)+fib_rec(m-2)
```

### Número m-ésimo de Fibonacci iterativamente

```python
def fib_it(m):
    var('i')
    L=[0, 1]
    i=2
    while i<=m:
        L.append(L[i-1] + L[i-2])
        i+=1
    return L[m]
    
```
