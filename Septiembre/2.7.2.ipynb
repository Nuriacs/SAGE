```python
var('a b i')
f(x,a,b) = x^4 - 6*x^2 + a*x + b
for i in range (0,1):
    g = f(a=0,b=i)
    show(g)
    show(plot(g, (x, -3, 3)))
```

```python
g = f(a=8,b=0)
show(g)
show(plot(g, (x, -3, 3)))
```

```python

g = f(a=8,b=0)
h = derivative(g)
p = derivative(h)
show(h)
show(plot(h, (x, -3, 3)))
show(p)
show(plot(p, (x, -3, 3)))
```
