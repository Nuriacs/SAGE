```python
'''La funcion guarda en una lista el resultado de la serie asociado
a cada limite superior entre 0 y un n dado'''
def suma(n):
    L = []
    suma = 0
    for i in srange(0, n+1):
        a = 2*i
        a = (factorial(a))^3
        a = a * (42*i + 5)
        a = a / (factorial(i)^6)
        a = a / (16^(3*i + 1))
        suma = suma + a
        if i > 0:
            L.append((i, 1/suma))
        
    return L

n = 100
show(line2d(suma(n)))


    
```
