### Ej 3

$ \text{Demostrar por inducción: }  \sum_{i = 1}^{i= n} n^3 = \frac{\left( n+1\right) ^2n^2}{4}$

```python
def f(n):
    return (n+1)**2*n**2/4

'''Caso base'''
print(1**3 == f(1))

'''Caso general'''
var('n')
print( ((n+1)**3 + f(n) - f(n+1))).expand()
```
